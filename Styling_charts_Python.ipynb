##**Let's give our tables some style using Python!**

####Objective: Generate insights from a 2016–2019 sales dataset for a major online department store in Argentina, and create visualizations to share key findings with other departments.

###**Preparation of the data**

```python
import pandas as pd

print(pd.__version__)
```

```python
sales = pd.read_csv('https://gist.githubusercontent.com/ahcamachod/d2efadda3f3833d65da750c64705ac42/raw/80aef6f3bad210d34b67ed85565f039a30e2d9e0/ventas.csv', sep='#')
sales.head()
```

```python
sales = sales[['fecha_pedido', 'modo_envio','nombre_cliente', 'segmento_cliente', 'ciudad',
              'provincia', 'region', 'departamento', 'tipo_producto',
              'ventas', 'cantidad', 'ganancia']]
sales.head(2)
```

```python
sales.shape
```

```python
sales.info()
```

####We only need to transform the date, as it's currently marked as an "object" but should be a datetime.

```python
sales['fecha_pedido'] = pd.to_datetime(sales['fecha_pedido'],format='%Y-%m-%d')
sales.head(2)
```

```python
sales.info()
```

####**Our DataFrame is now ready for analysis.**

###**Step 1: Identifying the top 10 clients.**

1. Let's create another dataframe named clients for this step.
2. with the nlargest function, python give us the 10 clients with more sales
3. since we want just a copy, we use the copy function
4. Reset index.

```python

clients = sales.groupby(['nombre_cliente'])['ventas'].sum().nlargest(10).copy()
clients = clients.reset_index()
clients
```

```python
clients.columns = ['Clients','Sales']
clients['Ranking'] = clients.index + 1
clients.set_index('Ranking', inplace = True)
clients
```

```python
#Since the previous tables weren't stored in any DataFrame, we’re now
#saving them in one named "style".
style = clients.style
style
```

```python
style.format({'Sales':'{:,.2f}'})
style
```

###**Step 2: Identifying the top products.**

```python
high_sales = sales.groupby(['tipo_producto'])[['ventas', 'ganancia']].sum()
```

```python
high_sales.index.name = 'tipo de producto'
high_sales
```

```python
product_style = high_sales.style
product_style
```

```python
product_style.format('USD {:,.2f}')
```

```python
product_style = product_style.format('USD {:,.2f}').highlight_max(color='lightgreen').highlight_min(color='red')
product_style
```

```python
product_style.background_gradient(cmap='Greens')
```

```python
#Let's change the titles for our table, to make it look better.
#For this we need to install CSS, so let's call it.

titles = {
        'selector':'th',
        'props':'font-weight:bold; font-family:Arial; text-align:center; text-transform:capitalize'
}
```

```python
product_style.set_table_styles([titles], overwrite=False)
```

##**Order distribution by region**

##**Where the majority of orders are concentrated within the country.**

```python
region = pd.DataFrame(sales['region'].value_counts())
region.columns = ['N° Orders']
region.index.name = 'Region'
region
```

```python
import numpy as np
```

```python
percentage = region['N° Orders'].to_numpy()
percentage = 100*percentage/percentage.sum()
percentage
```

```python
region['Orders Percentage'] = percentage
region
```

```python
#Let's fix the style
region_style =region.style
region_style
```

```python
#First headers and white table
headers = {
    'selector': 'th', #table head = th
    'props':'font-weight:bold; font-family:Arial; text-align:right; background-color: white;'
}

rows = {
    'selector': 'td',
    'props':'background-color: white;'
}

region_style.set_table_styles([headers,rows])
```

```python
region_style.format({'Orders Percentage':'{:.2f} %'}).bar(subset='Orders Percentage', vmin=0,vmax=100,color='#9cd3bb')
```

##**Identify the pattern per month**

```python
months = sales.copy()
months = months.sort_values('fecha_pedido')
months
```

```python
#take our df and create new attribute call month using string format time (strftime)

months['month'] = months['fecha_pedido'].dt.strftime('%Y - %b')
months = months.reset_index(drop=True)
months
```

```python
pd.set_option('display.max_columns', None)
```

```python
months_sales = months.pivot_table(index=months.departamento, columns='month', values='cantidad', aggfunc='sum', sort=False)
months_sales
```

```python
month_style = months_sales.style
month_style
```

```python
month_style.set_sticky(axis='index')
```

using CSS let's modofy it for good

```python
columns = {
    'selector':'.col_heading',
    'props':'font-weight:normal; font-family:Arial'
}

table = {
    'selector':'th,td',
    'props': 'text-align:left'
}

indixes = {
    'selector':'index_name',
    'props':'font-weight:bold; font-family:Arial ;text-align:right'
}

month_style.set_table_styles([columns, table, indixes], overwrite=False)
```

##**Performance Report**

```python
#Let's create new dt to prepera the data
df_client = sales.pivot_table(index='segmento_cliente',columns='modo_envio',values='ventas',aggfunc='sum')
df_client
```

```python
#total values for that we are gouing to creat total columns and total index
df_client['Total'] = df_client.sum(axis=1)
df_client.loc['Total'] = df_client.sum()
df_client
```

```python
#let's create a new style
sales_client = df_client.style.format('{:,.2f}')
sales_client
```

```python
#let's create our dictionary table

table2={
    'selector': 'td,th:not(.index_name)',
    'props':'font-weight:normal; font-family:Arial; text-align:center; background-color:white'
}

indixes2={
    'selector':'.index_name',
    'props':'font-weight:normal; font-style:italic; text-align:right; color:#696969'
}

sales_client.set_table_styles([table2,indixes2])
```

```python
#let's make some lines for our table
sales_client.set_table_styles({
    'Total':[
        {
            'selector':'th',
            'props':'border-top: 2px solid #181818'
        },
         {
            'selector':'td',
            'props':'border-top: 2px solid #181818'
        }
    ],
    'B2B':[
        {
            'selector': 'th',
            'props':'border-top: 2px solid #181818'
        },
         {
            'selector': 'td',
            'props': 'border-top: 2px solid #181818'
        }
    ]
}, overwrite=False, axis=1)
```
